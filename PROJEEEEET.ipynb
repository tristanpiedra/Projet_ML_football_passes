```python
#On calcule pour la première ligne du dataframe

def distance(df, sender, receveur) : 
    d = np.sqrt((df.iloc[0,3+sender]-df.iloc[0,3+receveur])**2 + (df.iloc[0,31+sender]-df.iloc[0,31+receveur])**2)
    return d
```
